```python
import os
import sys
import glob
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import multivariate_normal,beta,dirichlet,norm
from math import log,sqrt,pi
%matplotlib inline
plt.style.use('ggplot')
np.random.seed(1234)
from scipy.special import gamma
import ternary


np.set_printoptions(formatter={'all':lambda x: '%.3f' % x})
```

```python
p0 = norm(0,1)
alpha = 1000
vh = beta(1,alpha)
```

```python
p_x = []
p_y = []
breakoff = vh.rvs(size=1)
ratio = breakoff
product = 1-breakoff
while product>1e-10:
    print(f'product is: {product},ratio is: {ratio}')
    p_x.append(p0.rvs(size=1))
    p_y.append(ratio)
    breakoff = vh.rvs(size=1)
    ratio = breakoff*product
    product = product * (1-breakoff)
plt.figure()
plt.plot(p_x,p_y,'o') 

```
