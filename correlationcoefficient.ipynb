```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
event = np.ones(1000)
event[100:200]=5
event[400:500]=5
event[800:900]=5
```

```python
signal01 = np.ones(1000)
signal01[100:200]=100
signal01[400:500]=100
signal01[800:900]=100

signal02 = np.ones(1000)
signal02[100:200]=20
signal02[400:500]=20
signal02[800:900]=20

signal03 = np.ones(1000)
signal03[100:200]=200
signal03[400:500]=200
signal03[800:900]=0
```

```python
plt.figure()
plt.subplot(4,1,1)
plt.plot(event)
plt.title('event')
plt.subplot(4,1,2)
plt.plot(signal01)
plt.title('singal 01')
plt.subplot(4,1,3)
plt.plot(signal02)
plt.title('singal 02')
plt.subplot(4,1,4)
plt.plot(signal03)
plt.title('singal 03')
plt.show()
```

```python
std_e = np.std(event)
std01 = np.std(signal01)
std02 = np.std(signal02)
std03 = np.std(signal03)
print(std_e,std01,std02,std03)
```

```python
cov01 = np.cov(np.array([event,signal01]),bias=True)
cov02 = np.cov(np.array([event,signal02]),bias=True)
cov03 = np.cov(np.array([event,signal03]),bias=True)

print(cov01,cov02,cov03)
```

```python
r01=cov01/(std_e*std01)
r02=cov02/(std_e*std02)
r03=cov03/(std_e*std03)

print(r01,r02,r03)
```

```python
r01 = np.corrcoef(event,signal01)
r02 = np.corrcoef(event,signal02)
r03 = np.corrcoef(event,signal03)
```

```python
print(f'r01 is: {r01}')
print(f'r02 is:{r02}')
print(f'r03 is:{r03}')
```

```python
def cov(x,y): 
    mean_x = np.mean(x)
    mean_y = np.mean(y)

    dev_x = [i - mean_x for i in x]
    dev_y = [i - mean_y for i in y]

    numerator = sum([dev_x[i] * dev_y[i] for i in range(len(x))])/len(dev_x)

    denominator_x = np.sqrt(sum([i**2 for i in dev_x])/len(dev_x))
    denominator_y = np.sqrt(sum([i**2 for i in dev_y])/len(dev_y))
    corr_coef = numerator / (denominator_x * denominator_y)
    print("Correlation Coefficient:", corr_coef)
    print(f'the covariance is:{numerator}')
    print(f'the std of x is:{denominator_x}')
    print(f'the std of y is:{denominator_y}')
```
